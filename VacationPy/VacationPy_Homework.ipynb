### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
#Set GMAP with heat layer
gmaps.configure(api_key=g_key)

City_Locations = cities_file[["Latitude","Longitude"]]

humidity = cities_file['Humidity'].astype(float)

fig = gmaps.figure(zoom_level=2, center=(0,0))

heat_layer = gmaps.heatmap_layer(City_Locations, weights=humidity, 
                                 dissipating=False, max_intensity=100,
                                 point_radius=1)

fig.add_layer(heat_layer)

fig
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
#add Hotel Name column to original DF

cities_file['Hotel Name'] = ""

Ideal_Vacation = cities_file[(cities_file['Max Temp (F)']<80) & (cities_file['Country']=="ES")]

hotel_df = Ideal_Vacation

radius_amount = 5000


params = {
    "radius": radius_amount,
    "type": "lodging",
    "key": g_key,}

#iterate through each lat/long combination based off city results

for index, row in hotel_df.iterrows():
    
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
        
    lat_codes = row['Latitude']
    lng_codes = row['Longitude']
    
    params['location']= f'{lat_codes}, {lng_codes}'
    
    response = requests.get(base_url, params=params)
    
    places_data = response.json()
    
    
      
    hotel_df.loc[index, 'Hotel Name'] = places_data['results'][0]['name']
```

```python
#check our work
hotel_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Latitude", "Longitude"]]
```

```python
# Add marker layer ontop of heat map

mark_layer = gmaps.marker_layer(locations, info_box_content = hotel_info)

fig.add_layer(mark_layer)



# Display figure

fig
```
